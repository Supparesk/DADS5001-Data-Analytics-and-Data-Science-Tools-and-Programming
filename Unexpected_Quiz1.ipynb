```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-poster')
%matplotlib inline
```

```python
df = pd.read_csv('/content/sample_data/company_sales_data.csv')
df.head()
```

# 1. Read Total profit of all months and show it using a line plot

image.png

```python
df.plot(kind = 'line', x = 'month_number', y = 'total_profit', color = 'blue',
       title = 'Country Profit per month',
       ylabel =  'profit',
        xlabel = 'Month number',
        grid = True,
       legend = False)
```

#2. Get total profit of all months and show line plot with the following Style properties

image.png

```python
df.plot(kind = 'line', x = 'month_number', y = 'total_profit', 
        color = 'red', 
        ls = '--', marker = 'o',
        markerfacecolor = 'black',
       lw = 3,
        markersize = 8,
         ylabel =  'profit',
        xlabel = 'Month number',
       title = 'Company Sale last Year')
```

#3. Read all product sales data and show it  using a multiline plot

image.png

```python
df.plot(kind = 'line', x = 'month_number', y = ['facecream', 'facewash', 'toothpaste', 'bathingsoap',
       'shampoo', 'moisturizer'], ls = '--', marker = 'o',
       ylabel =  'Sale unit in number',
        xlabel = 'Month number',
       figsize = (15,8))
```

# 4. Read toothpaste sales data of each month and show it using a scatter plot

image.png

```python
df.plot(kind = 'scatter', x = 'month_number', y = ['toothpaste'],grid = True,
       s = 40)
```

# 5. Read face cream and face wash product sales data and compare them using the bar chart

image.png

```python
df.plot(kind = 'bar', x = 'month_number', y = ['facecream', 'facewash'], grid = True, figsize = (10,5))
```

# 6. Read sales data of bathing soap of all months and show it using a bar chart.

image.png

```python
df.plot(kind = 'bar', x = 'month_number', y = ['bathingsoap'], rot = 360)
```

# 7. Read the total profit of each month and show it using the histogram to see the most common profit ranges

image.png

```python
df.total_profit.plot(kind = 'hist', alpha = 0.5)
```

# 8. Calculate total sale data for each product and show it using a Pie chart

image.png

```python
df[['facecream', 'facewash', 'toothpaste', 'bathingsoap',
       'shampoo', 'moisturizer']].sum().plot(kind = 'pie', autopct='%.1f%%',
        figsize = (10,5)) #legend = True, label = False)
```

# 9. Read Bathing soap and facewash of all months and display it using the Subplot

image.png

```python
fig, ax = plt.subplots(2,1, sharex = True, figsize = (10,4))
df.plot(kind = 'line', x = 'month_number', y = 'bathingsoap', color = 'red', ls = '--', marker = 'o', ax = ax[0])
df.plot(kind = 'line', x = 'month_number', y = 'facewash', color = 'green', ls = '--', marker = 'o', ax = ax[1])
```

# 10. Read all product sales data and show it using the area plot

image.png

```python
df.plot(kind = 'area', x = 'month_number', y = ['facecream', 'facewash', 'toothpaste', 'bathingsoap',
       'shampoo', 'moisturizer'],
        alpha = 0.7,
       figsize =(12,5),
       ylabel = "sale unit number")
```
